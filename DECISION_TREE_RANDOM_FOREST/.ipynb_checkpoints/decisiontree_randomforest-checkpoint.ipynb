## How to Approach a Machine Learning Problem:

1. Explore the data and find correlations between inputs and targets
2. Pick the right model, loss functions and optimizer for the problem at hand
3. Scale numeric variables and one-hot encode categorical data
4. Set aside a test set (using a fraction of the training set)
5. Train the model
6. Make predictions on the test set and compute the loss

## Preparing the Data for Training

We'll perform the following steps to prepare the dataset for training:

1. Create a train/test/validation split
2. Identify input and target columns
3. Identify numeric and categorical columns
4. Impute (fill) missing numeric values
5. Scale numeric values to the $(0, 1)$ range
6. Encode categorical columns to one-hot vectors


```python
# Since data is ordinal (i.e based on date, data will be split based on date)

# Convert string object datatype to datetime
year = pd.to_datetime(df["Date"]).dt.year

# Plot count of date in terms of years
plt.title("Count of years in data")
sns.countplot(x=year)
plt.ylabel("Observations")
```

### Create a train/test/validation split

```python
### Split data into training, validation and testing...since data is ordinal (i.e has order based on date), the split is based on date

# We set data earlier than 2015 to train model
train_df = df[year < 2015]
# Data for 2015 is to validate
val_df = df[year == 2015]
# Data for dates beyond 2015 will be used for testing
test_df = df[year > 2015]
```

```python
# Inspect shape of split data

total_rows = df.shape[0]

print(f"Shape of training data: {train_df.shape} : {round(100*train_df.shape[0]/total_rows)}%")
print(f"Shape of validation data: {val_df.shape} : {round(100*val_df.shape[0]/total_rows)}%")
print(f"Shape of testing data: {test_df.shape} : {round(100*test_df.shape[0]/total_rows)}%")
```

```python
train_df.head()
```

### Identify input (features) and target (labels) columns



```python
# Binary classification column
target_cols = "RainTomorrow"

# Take all columns from 1st index to last index(not-inclusive)
input_cols = list(train_df.columns)[1:-1]
```

```python
input_cols
```

```python
# Establish training, val and testing features and labels by making copies of df so that the original will not be disturbed

train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_cols].copy()

val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_cols].copy()

test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_cols].copy()
```

### Identify numeric and categorical features

```python
# Return a subset of the DataFrame's columns based on the column dtypes for training features

numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
print("These are the categorical columns")
```

```python
# Get categorical data
categorical_cols = train_inputs.select_dtypes("object").columns.tolist()

print("These are the numeric columns")
numeric_cols
```

### Impute (fill) missing numeric values

```python
# Check for null values in dataset
df.isna().sum().sort_values(ascending=False)
```

```python
from sklearn.impute import SimpleImputer

# Create instance of imputer based on mean of the columns
imputer = SimpleImputer(strategy='mean')

# Fit imputer to main data
imputer.fit(df[numeric_cols])
```

```python
help(SimpleImputer())
```

```python
# Check mean of each numeric column of raw df

list(imputer.statistics_)
```

```python
# Assign training, validation and testing inputs with tranformation of their numeric columns via imputation

train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
# Confirm if there are any missing values after imputation

train_inputs[numeric_cols].isna().sum()
```

### Scale numeric values to the $(0, 1)$ range

```python
from sklearn.preprocessing import MinMaxScaler

help(MinMaxScaler)
```

```python
# Create instance of feature scaler
scaler = MinMaxScaler()

# Fit scaler to raw data
scaler.fit(X=df[numeric_cols])
```

```python
# Transform training features of numerical columns to be in range of 0 and 1

train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
# Check new range of data after transformation
# The values of feature ranges should now be in range of 0 and 1

train_inputs[numeric_cols].describe().loc[["min", "max"]]
```

### Encode categorical columns to one-hot vectors

machine learning model can only work with numerical data, need to encode categorical data into a numeric format

```python
# Check quantity of null values in categorical columns

df[categorical_cols].isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create instance of encoder to  process each categorical value into separate columns

# ignore if an unknown categorical feature is present during transformation. 
# Resulting one-hot encoded columns for this feature will be all zeros
encoder = OneHotEncoder(sparse=False, handle_unknown = 'ignore')

# Fit one hot encoder to raw df of categorical columns
encoder.fit(df[categorical_cols])
```

```python
# See the list of generated categories from the encoder
encoder.categories_
```

```python
# Generate list of encoded columns by using the .get_feature_names_out() method

# Encoded array
encoded_array = encoder.get_feature_names_out(categorical_cols)

# Convert array to list
encoded_cols = list(encoded_array)
encoded_cols
```

```python
# Assign each categorical column of training, validation and testing by transforming their columns with onehotencoding
 
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

### Hyperparameter tuning -  "max tree depth"

```python
# Create helper function to see the effect of different hyperparameters

def check_error(max_tree_depth):
    model = DecisionTreeClassifier(max_depth=max_tree_depth, random_state=40)
    # Fit model to training data
    model.fit(X=X_train, y=train_targets)
    # Compute mean accuracy by obtaning error (cost) on training data.
    train_accuracy = model.score(X=X_train, y=train_targets)
    train_error = 1 - train_accuracy
    # Compute mean accuracy by obtaning error (cost) on validation data.
    val_accuracy = model.score(X=X_val, y=val_targets) 
    val_error = 1 - val_accuracy
    
    # Return dictionary of max depth, training error and validation error
    
    return {"Max depth": max_tree_depth, 
            "Training error": train_error, 
            "Validation error": val_error}
    
```

```python
# Create dataframe with range of different tree depths

# Use list comprehension
evaluation_df = pd.DataFrame([check_error(max_tree_depth) for max_tree_depth in range(1, 21)])
evaluation_df
```

```python
# Plot cost function curve

plt.figure(figsize=(15, 7))
plt.plot(evaluation_df["Max depth"], evaluation_df["Training error"])
plt.plot(evaluation_df["Max depth"], evaluation_df["Validation error"])
plt.xlabel("Max Tree Depth")
plt.xticks(range(0, 21, 1))
plt.ylabel("Cost (Error)")
plt.title("Training and Validation cost vs max tree depth")
plt.legend(["Training", "Validation"])
```

From the figure above, the max tree depth with lowest cost for the validation data is ~7. As number of max tree depth increases, the trend of training and validatin error is inverse. i.e training error decreases due to overfitting 

```python
# Apply decision tree with the hyperparameter of max tree depth adjust to 7
model = DecisionTreeClassifier(max_depth=7, random_state=40)

# Fit new model to training data
model.fit(X_train, train_targets)

# Compute modelscore on validation data
model.score(X_val, val_targets)
```

## Hyperparameter tuning of Random Forest

### 1) Adjust the n_estimators hyperparameter

- Controls number of decision trees to be used in random forest. Default is 100.
- Better to use as few as possible
- For larger dataset, more n_estimators may be helpful

#### Demo with n_estimators  of 50 estimators

```python
# Create a model with the adjust hyperparameter
model = RandomForestClassifier(random_state=40, n_estimators=50)
# Fit model on training data
model.fit(X_train, train_targets)
# Compute accuracy of model on training and validation data
model.score(X_train, train_targets), model.score(X_val, val_targets)
```

### 1) Adjust max_depth and max_leaf_nodes hyperparameters

```python
# Create helper function to generater the testing parameters

# Let the model accepts multiple key word arguments kwargs
def test_parameters(**hyperparameters):
    # Intantiate model
    model = RandomForestClassifier(random_state=40, n_jobs=-1, **hyperparameters)
    # Fit model to training data
    model.fit(X_train, train_targets)
    # Return the accuracy in form of a tuple for both training and val data
    return model.score(X_train, train_targets), model.score(X_val, val_targets)
    
```

```python
# Call helper function with different custom hyperparameters

test_parameters(max_depth=15)
```

```python
test_parameters(max_leaf_nodes=2**5)
```

```python
test_parameters(max_leaf_nodes=2**20)
```

### 2) Adjust min_sample_split and min_samples_leaf hyperparameters


By default, the decision tree classifier tries to split every node that has 2 or more rows/samples. You can increase the values of these arguments to change this behavior and reduce overfitting, especially for very large datasets.

```python
test_parameters(min_samples_split=3, min_samples_leaf=2)
```

```python
test_parameters(min_samples_split=100, min_samples_leaf=60)
```

### 3) Adjust min_impurity_decrease hyperparameters

This argument is used to control the threshold for splitting nodes. A node will be split if this split leads to a decrease of the impurity (Gini index) - cost function greater than or equal to this value. It's default value is 0, and you can increase it to reduce overfitting.

```python
 test_parameters(min_impurity_decrease=1e-6)
```

```python
test_parameters(min_impurity_decrease=1e-2)
```

### 4) Adjust bootstrap and max_samples hyperparameter

By default, a random forest doesn't use the entire dataset for training each decision tree. Instead it applies a technique called **bootstrapping**. For each tree, rows from the dataset are picked one by one randomly, with replacement i.e. some rows may not show up at all, while some rows may show up multiple times.

if bootstrap argument is passed as **False**, entire dataset is used in the decision tree. By default, it is set to **True**



```python
test_parameters(bootstrap=False)
```

When bootstrapping is enabled, you can also control the number or fraction of rows to be considered for each bootstrap using max_samples. This can further generalize the model. By default it is set to **None**

```python
# Example setting 85% of max samples
test_parameters(max_samples=0.85)
```

```python
# Example setting 50%
test_parameters(max_samples=0.50)
```

### 5) Adjust class weight hyperparameter

```python
# Check possible classification targets
model.classes_
```

```python
# Check the fraction of each binary classificatin target

train_targets.value_counts()/len(train_targets)
```

class_weight is the Weights associated with classes in the form ``{class_label: weight}``.
    If not given, all classes are supposed to have weight one.

    The "balanced" mode uses the values of y to automatically adjust weights

```python
# Use the 'balanced' default argument for class_weight to autmatically assign weightage to each classifier target
test_parameters(class_weight='balanced')
```

```python
# Assign 1:2 ratio for class weight i.e the "Yes" classfication target assigned weight of 2 since the number of samples are less than "No"
test_parameters(class_weight={"No":1, "Yes":2})
```

```python
# Helper function to produce prediction on single imput
def generate_predictions(input_data):
    input_df = pd.DataFrame([input_data])
    # Impute missing values and scale numerical columns of df
    input_df[numeric_cols] = imputer.transform(input_df[numeric_cols])
    input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])
    # Insert encoded categorical cols into df
    input_df[encoded_cols] = encoder.transform(input_df[categorical_cols] )
    # Define input features, X
    X = input_df[numeric_cols + encoded_cols]
    # Generate predictions
    prediction = model.predict(X)[0]
    # Generate probability of predictions
    probability = model.predict_proba(X)[0][list(model.classes_).index(prediction)]
    return prediction, probability
```
